```python
#11. Categorise the track with their corresponding artists based on the views in YouTube
Dt=df[['Track','Artist','YouTube Views']]
Dt['Remarks']=np.nan
Dt['Remarks'][Dt['YouTube Views']<5000000]='Less viewed'
Dt['Remarks'][(Dt['YouTube Views']>=5000000) & (Dt['YouTube Views']<500000000)]='Average'
Dt['Remarks'][(Dt['YouTube Views']>=500000000) & (Dt['YouTube Views']<5000000000)]='Highly viewed'
Dt['Remarks'][Dt['YouTube Views']>=5000000000]='Most viewed'
Dt.sort_values(by='YouTube Views',ascending=False)
```

```python
#14. Most popular songs in TikTok
TikTok_Views=df[['Artist','Track','TikTok Views']].sort_values(by='TikTok Views',ascending=False).head(10)
TikTok_Posts=df[['Artist','Track','TikTok Posts']].sort_values(by='TikTok Posts',ascending=False).head(10)
TikTok_Popularity=TikTok_Views.merge(TikTok_Posts)
TikTok_Popularity['Popularity']=TikTok_Popularity['TikTok Views']+TikTok_Popularity['TikTok Posts']
TikTok_Popularity[['Artist','Track','Popularity']].sort_values(by='Popularity',ascending=False)
```
